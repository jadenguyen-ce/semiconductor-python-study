```python
# Load the CSV
import pandas as pd
tool_log =  pd.read_csv("C:/Users/loanp/OneDrive/Desktop/JADE/ENGR_6MPLAN/python/week_2/tool_log.csv")


# Inspect it with at least 3 different methods
tool_log
tool_log.shape
tool_log.dtypes

# Select a subset of columns
tool_log.issue_description
tool_log[['issue_description', 'frequency']]


# Filter rows based on a condition
urgent = tool_log[ tool_log['frequency'] >= 3] # rows that have issue happened more than once
urgent

# Group by tool name and summarize a numeric column
down_tool = tool_log.groupby('tool_name')['downtime_hours'].sum() #result the total downtime of each tool
down_tool

# Check for and handle missing values
missing_data = tool_log[pd.isnull(tool_log.notes)]
missing_data

# Plot a histogram and a bar chart with proper labels
import matplotlib.pyplot as plt
plot = plt.hist(tool_log['downtime_hours'])
```
