```python
import pandas as pd

tool_log = pd.read_csv("C:/Users/loanp/OneDrive/Desktop/JADE/ENGR_6MPLAN/python/week_2/tool_log.csv")
print (tool_log)
```

```python
import matplotlib.pyplot as plt
# Plot a histogram of the downtime or cycle time column
plt.hist(tool_log.downtime_hours)
plt.title('Histogram of the downtime')
plt.show()

# Plot a scatter plot (plot with points only) of two numeric columns against each other
plt.plot(tool_log.downtime_hours, tool_log.frequency, 'bo')
plt.xlabel("Down time hours")
plt.ylabel("Down time Frequency")
plt.title("Frequency per Downtime")
plt.show()

# Try a bar chart showing issue count per tool (tool_name, frequency)
to_plot = tool_log.groupby('tool_name')['frequency'].sum()
plt.bar(to_plot.index, to_plot.values)
# plt.xticks(rotation=45)
plt.xlabel("Tool Name")
plt.ylabel("Total Issue Frequency")
plt.title("Issue Frequency per Tool")
plt.show()

# Add proper axis labels and a title to each plot

```
