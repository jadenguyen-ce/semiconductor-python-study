```python
import pandas as pd 
# Load dataset
df = pd.read_csv("C:/Users/loanp/OneDrive/Desktop/JADE/ENGR_6MPLAN/python/week_1/fake_metrology.csv.txt")
print (df)

# Basic summary stats on a measurement column
print(f"Median: {df["thickness_nm"].median(): .2f}")
mean = df["thickness_nm"].mean()
# std stands for Standard Deviation — it measures how spread out your values are around the mean.
std  = df["thickness_nm"].std()

print(f"Mean: {mean:.2f}")
print(f"Std:  {std:.2f}")

# 3-sigma rule: ~99.7% of data falls within mean ± 3×std
upper_range = mean + 3*std
lower_range = mean - 3*std
print(f"Expected range (3-sigma): {lower_range: .2f} to {upper_range: .2f}")

# Use map() to normalize or flag values
df["flagged"] = df["thickness_nm"].map(lambda x: "out of spec" if x < lower_range or x > upper_range else "ok")
print(df["flagged"].value_counts())

```
